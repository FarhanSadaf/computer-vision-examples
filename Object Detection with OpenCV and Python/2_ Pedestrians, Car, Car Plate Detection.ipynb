### Pedestrians Detection from video 

```python
import cv2

# Load the video
cap = cv2.VideoCapture('../Videos/pedestrians-walking 2.mp4')

# Load cascade
body_cascade = cv2.CascadeClassifier('../Haarcascades/haarcascade_fullbody.xml')

while cap.isOpened():
    ret, frame = cap.read()
    
    if ret:
        bodies = body_cascade.detectMultiScale(frame, 1.1, 3)
        
        for (x, y, w, h) in bodies:
            cv2.rectangle(img=frame, pt1=(x, y), pt2=(x+w, y+h), color=(0, 255, 0), thickness=2)
        
        cv2.imshow('Pedestrians Detection', frame)
        
        if cv2.waitKey(1) & 0xFF == 27:
            break
    else:
        break

cap.release()
cv2.destroyAllWindows()
```

### Car Detection from video

```python
import cv2

# Load the video
cap = cv2.VideoCapture('../Videos/cars 1.mp4')

# Load cascade
car_cascade = cv2.CascadeClassifier('../Haarcascades/haarcascade_car.xml')

while cap.isOpened():
    ret, frame = cap.read()
    
    if ret:
        cars = car_cascade.detectMultiScale(frame, 1.1, 2)
        
        for (x, y, w, h) in cars:
            cv2.rectangle(img=frame, pt1=(x, y), pt2=(x+w, y+h), color=(0, 255, 0), thickness=2)
        
        cv2.imshow('Pedestrians Detection', frame)
        
        if cv2.waitKey(1) & 0xFF == 27:
            break
    else:
        break

cap.release()
cv2.destroyAllWindows()
```

### Car Plate Detection from image

```python
import cv2
import matplotlib.pyplot as plt

# Load the image
img = cv2.imread('../Images/car-plate 4.jpg')

# Load cascade
plate_cascade = cv2.CascadeClassifier('../Haarcascades/haarcascade_russian_plate_number.xml')

plates = plate_cascade.detectMultiScale(img)

for (x, y, w, h) in plates:
    cv2.rectangle(img=img, pt1=(x, y), pt2=(x+w, y+h), color=(0, 255, 0), thickness=3)

# Show image
plt.figure(figsize=(10, 6))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```
