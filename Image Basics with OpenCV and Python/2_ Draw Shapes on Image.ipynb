```python
import numpy as np
import matplotlib.pyplot as plt

import cv2
```

```python
black_img = np.zeros((512, 512, 3), dtype=np.int16)

plt.imshow(black_img)
```

### Draw a Circle

```python
cv2.circle(img=black_img, 
          center=(400, 100),
          radius=50,
          thickness=8,
          color=(255, 0, 0))

plt.imshow(black_img)
```

### Draw a Filled Circle

```python
cv2.circle(img=black_img, 
          center=(400, 190),
          radius=40,
          thickness=-1,
          color=(0, 255, 0))

plt.imshow(black_img)
```

### Draw a Rectangle

```python
cv2.rectangle(img=black_img,
             pt1=(100, 200),
             pt2=(200, 300),
             thickness=6,
             color=(255, 0, 0))

plt.imshow(black_img)
```

### Draw a Filled Rectangle

```python
cv2.rectangle(img=black_img,
             pt1=(100, 300),
             pt2=(200, 400),
             thickness=-1,
             color=(0, 255, 0))

plt.imshow(black_img)
```

### Draw Triangle

```python
vertices = np.array([
    [350, 300],
    [300, 400],
    [500, 400]
])

pts = vertices.reshape(-1, 1, 2)

cv2.polylines(img=black_img,
             pts=[pts],
             isClosed=True,
             thickness=5,
             color=(0, 0, 255))

plt.imshow(black_img)
```

### Draw Filled Triangle

```python
vertices = np.array([
    [150, 0],
    [100, 100],
    [200, 100]
])

pts = vertices.reshape(-1, 1, 2)

cv2.fillPoly(img=black_img,
             pts=[pts],
             color=(255, 161, 201))

plt.imshow(black_img)
```

### Draw a Line

```python
cv2.line(img=black_img,
        pt1=(220, 480),
        pt2=(330, 20),
        thickness=3,
        color=(255, 255, 255))

plt.imshow(black_img)
```

### Write Text

```python
text = 'Rhyme'

cv2.putText(img=black_img,
           text=text,
           org=(400, 480),
           fontFace=cv2.FONT_HERSHEY_SIMPLEX,
           fontScale=1,
           color=(255, 255, 255),
           thickness=2,
           lineType=cv2.LINE_AA)

plt.imshow(black_img)
```
